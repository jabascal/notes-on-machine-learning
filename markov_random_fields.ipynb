#Theory

Directed graphs, also called Markov networks or Markov Random Fields (MRF) are graphical models where random variables are represented by nodes that are connected with undirected edges. In contrast to directed graphs (Bayesian networks) where directed edges encapsulate causality relations, in MRF influence flows on both directions. 

For MRF, factors are not restricted to probability distributions (marginals or CPDs) but defined as positive potentials over their variable scope. The joint distribution is built as the factor product of unnormalized potentials which are then normalized by dividing it by the partition function.

The unnormalized factorization is given by 

$$
\tilde{P}(A,B,C,D)=\Psi(A,B)\Psi(B,C)\Psi(C,D)\Psi(D,A),
$$

then the joint distribution is obtained by normalization

$$
\tilde{P}(A,B,C,D)=\frac{1}{Z}\tilde{P}(A,B,C,D),
$$

where $Z$ if the partition function gives as

$$
Z=\sum_{A,B,C,D} \tilde{P}(A,B,C,D).
$$

Normalization is one of the limitations of MRF, as it requires to sum up over all possible states (although it is not needed for local conditional conditional distributions or for marginals, which can be normalized at the end).
